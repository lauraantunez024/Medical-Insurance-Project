# California Kindercare violations 

#### For this project, the journalist wanted to know how many times specific citation numbers appeared in the California data set. Prior to the running the python scripts, I ommited some empty columns at the end of the original file that were preventing python from running on the script. I start by setting up my environment and adding in the packages I might use during my analysis.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
import regex
from collections import Counter
from itertools import chain
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)


df = pd.read_csv('Childcareclean2.csv')
print(df.head)
```

### The data is messy and filled with columns that will not be used. I alter the data frame to include only relevant information.

```python
data = df[['Facility Type' ,
           'Facility Number',
           'Facility Name',
           'Licensee',
           'Facility State',
           'Facility Zip',
           'County Name',
           'Regional Office',
           'Facility Capacity',
           'Facility Status',
           'License First Date',
           'Closed Date',
           'Last Visit Date',
           'Inspection Visits',
           'Complaint Visits',
           'Other Visits',
           'Total Visits',
           'Citation Numbers']]
print(data.head())





```

### In this case, I thought the best course of action was to add columns to the end of the table that counted the specific citation numbers we were looking for. The journalist also wanted to include citations that might have numbers or letters appended at the end. The regex script catches scripts that begin with the specific citation numbers, but will also count appended letters and numbers. 

```python
data['101223(a)(3) count'] = df['Citation Numbers'].str.count(r'101223[\.\(\s]?a[\.\)\s]?(\([12]\)){0,2}[\.\(\s]?3[\.\)\s]?')
data['101416.5 variation counts'] = df['Citation Numbers'].str.count(r'101416\.5([\(\s]\w[\)\s])?([\(\s]\w[\(\s])?([\(\s]\w[\)\s])?([\(\s]\w[\)\s])?([\(\s]\w[\)\s])?([\(\s]\w[\)\s])?')
data['101216.3(a) count'] = df['Citation Numbers'].str.count(r'101216\.3\(a\)([\(\s]\w[\)\s])?([\(\s]\w[\(\s])?([\(\s]\w[\)\s])?([\(\s]\w[\)\s])?([\(\s]\w[\)\s])?([\(\s]\w[\)\s])?')
```

### Once columns with appropriate counts have been added, I sort by the values of those columns to make sure everything that is captured is accurate.

Sorted by 101416.5 variations:

```python
x= data.sort_values(by=['101416.5 variation counts'], ascending=False)

print(x.head())
```

Sorted by 101223(a)(3):

```python
y = data.sort_values(by=['101223(a)(3) count'], ascending=False)
print(y.head)
```

Sorted by 101216.3(a):

```python
z = data.sort_values(by=['101216.3(a) count'], ascending=False)
print(z.head())
```
